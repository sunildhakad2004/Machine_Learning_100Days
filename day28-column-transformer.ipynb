```python
import pandas as pd
```

```python
data = pd.read_csv(r"Datasets\train.csv")
data
```

```python
data.info()
```

```python
for column_name, none_values in data.isnull().sum().items():
    print(f'{column_name}: {none_values}')
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(data.drop(columns=['Survived'])
                                                 ,data['Survived'],test_size=0.1)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.impute import SimpleImputer

si = SimpleImputer()

X_train_value = si.fit_transform(X_train[['Age']])

X_test_value = si.fit_transform(X_test[['Age']])

X_train_value.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder

ode = OrdinalEncoder(categories=[['female','male']])

X_train_sex = ode.fit_transform(X_train[['Sex']])
X_test_sex = ode.transform(X_test[['Sex']])

X_train_sex.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse_output=False)

X_train_Embarked = ohe.fit_transform(X_train[['Embarked']])

X_test_Embarked = ohe.transform(X_test[['Embarked']])

X_train_Embarked.shape
```

```python
import numpy as np
```

```python
X_train_transformed = np.concatenate((X_train_sex,X_train_Embarked,X_train_value),axis=1)

X_test_transformed = np.concatenate((X_test_sex,X_test_Embarked,X_test_value),axis=1)

X_train_transformed.shape
```

```python
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['Age']),
    ('tnf2',OrdinalEncoder(categories=[['female','male']]),['Sex']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['Embarked'])
],remainder='drop')
```
