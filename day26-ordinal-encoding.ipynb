```python
import pandas as pd

data = pd.read_csv("Datasets\customer.csv")
print(data.head(5))
```

```python
print(data.iloc[:,2:])
```

```python
X = data[['review','education']]
y = data['purchased']

X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import OrdinalEncoder

ode = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
ode.fit(X_train)
```

```python
X_train = ode.transform(X_train)
X_train
```
